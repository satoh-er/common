# 換気について

### 換気とは
**換気**：室内の空気を追い出し、室外の空気を取り入れること<br>
**通風**：居住者が涼しさを得るために風を取り入れること

### 換気の種類
自然の力(室内外の圧力や温度差)を利用する換気方法を**自然換気**といい、ファン等の機械で換気する方法を**機械換気**という。


### 自然換気の種類

### 風力換気
風の圧力差で換気すること。図1 のように風上で空気が壁に当たったとき、圧力は上がり正圧になる。風下では負圧になり、引っ張られて風が通る。<br>
<br>
<div align="center"><img src="furyoku_kanki.png" alt="alt テキスト" title="風力換気" width="30%"></div>

$$ 図1　風力換気 $$
<br>

換気量 $Q$ を式1に示す。
$$ Q = A \cdot v \tag{1}$$
風速 $v$を、空気密度 $\rho$ [kg/m<sup>3</sup>]、風上、風下側の圧力差 $\Delta p$ [Pa] で表すと
$$ v = \alpha \sqrt{\frac{2}{\rho}\Delta p} \tag{2}$$
となり式1に代入すると

$$ Q = \alpha \cdot A \sqrt{\frac{2}{\rho}\Delta p}  \tag{3}$$
となる。
$$ p= C \frac{\rho}{2} v^{2}$$
より
$ \Delta p$は風上、風下側の風圧係数 $C_1$、$C_2$ で表すと 
$$ \Delta p = C_1 \frac{\rho}{2} v^{2} - C_2 \frac{\rho}{2} v^{2} \tag{4}$$
式4を式3に代入すると
$$ Q = \alpha \cdot A \cdot v \cdot \sqrt{C_1 - C_2} \tag{5}$$
となる。

換気量 $Q$ [m<sup>3</sup>/s] $\times$ 3600 = [m<sup>3</sup>/h]、流量係数 $\alpha$、開口部面積 $A$ [m<sup>2</sup>]、風速 $v$ [m/s]、風上側風圧係数 $C_1$、風下側風圧係数 $C_2$



### 温度差換気 (重力換気)
室内外の温度差を利用して換気すること。例えば、冬の場合、気温は室外 ＜ 室内となり室外の空気は密度が高く、下降する。室内の空気は膨張して密度が低く、上昇する(図2-1)。
<br>
<div align="center"><img src="ondosa_kanki.png" alt="alt テキスト" title="温度差換気" width="25%"></div>

$$ 図2-1　温度差換気 $$
<br>

換気量 $Q$は上の式3より
$$ Q = \alpha \cdot A \sqrt{\frac{2}{\rho_{o}}\Delta p}  \tag{3}$$
<!-- 下部の開口部の室内外の圧力差を $(\rho_{o} - \rho_{i}) \cdot g h_1$、上部の開口部の室内外の圧力差を $- (\rho_{o} - \rho_{i}) \cdot g h_2$ とすると<br> -->
下部(流入口)と上部(流出口)との圧力差 $\Delta p$ を室外の空気の密度 $\rho_{o}$、室内の空気の密度 $\rho_{i}$との差、上部、下部（流入、流出口）の高低差 $h$で表すと
$$ \Delta p = (\rho_{o} - \rho_{i}) gh \tag{6}$$
<!-- $$ \Delta p = (\rho_{o} - \rho_{i}) \cdot g h_1 - (- (\rho_{o} - \rho_{i}) \cdot g h_2) = (\rho_{o} - \rho_{i}) \cdot (h_{1} + h_{2}) \cdot g$$ 
ここで、$h = h_{1} + h_{2}$とすると上式は
$$ \Delta p = (\rho_{o} - \rho_{i}) \cdot g h \tag{6}$$ -->
となり、式6を式3に代入すると
$$ Q = \alpha \cdot A \sqrt{\frac{2gh(\rho_{o} - \rho_{i})}{\rho_o}}  \tag{7}$$
となる。$\rho_{o} (t_{o} + 273) = \rho_{i} (t_{i} + 273) $ の関係より、$\rho_{o}$、$\rho_{i}$ を消去すると換気量 $Q$は

$$ Q = \alpha \cdot A \cdot \sqrt{\frac{2gh(t_i -t_o)}{273 + t_i}} \tag{8}$$
となる。

換気量 $Q$ [m<sup>3</sup>/s] $\times$ 3600 = [m<sup>3</sup>/h]、流量係数 $\alpha$ 、開口部面積 $A$ [m<sup>2</sup>]、重力加速度 $g$ [m/s<sup>2</sup>]、流入口と流出口との高低差 $h$ [m]、室内温度 $t_i$[℃]、室外温度 $t_o$[℃]

### 参考　相当開口面積 (開口部合成)の導出

#### 開口部が直列のとき
上部と下部の風量は同じで、圧力差の総和 $\Delta p$は 下部の圧力差 $\Delta p_{1}$、上部の圧力差 $\Delta p_{2}$との和であるので、
$$ \Delta p = \Delta p_{1} + \Delta p_{2} = \frac{\rho}{2} \left(\frac{Q}{\alpha_{1}A_{1}}\right)^{2} + \frac{\rho}{2} \left(\frac{Q}{\alpha_{2}A_{2}}\right)^{2}$$
上式を書き換えると開口部が直列のとき、相当開口面積 (開口部合成) $\alpha A$ は
$$ \left(\frac{1}{\alpha A}\right)^{2} = \left(\frac{1}{\alpha_{1}A_{1}}\right)^{2} + \left(\frac{1}{\alpha_{2}A_{2}}\right)^{2} $$
$$ \alpha A = \frac{1}{\sqrt{\left(\frac{1}{\alpha_{1}A_{1}}\right)^{2} + \left(\frac{1}{\alpha_{2}A_{2}}\right)^{2}}} $$


#### 開口部が並列のとき
上部と下部の圧力差は同じで、風量の総和 $Q$ は下部 $Q_{1}$ と上部 $Q_{2}$ の和であるので、
$$ Q = Q_{1} + Q_{2} = \alpha_{1}  A_{1} \sqrt{\frac{2}{\rho}\Delta p} + \alpha_{2}  A_{2} \sqrt{\frac{2}{\rho}\Delta p} $$

上式を書き換えると開口部が並列のとき、相当開口面積 (開口部合成) $\alpha A$ は
$$ \alpha A = \alpha_{1}  A_{1} + \alpha_{2}  A_{2} $$
となる。

### 必要換気量と換気回数<br>

**必要換気量**：室内の汚染物質を許容濃度以下の維持するための必要な換気量をいう。<br>
**換気回数**：1時間に何回換気されるかをいう。

換気回数 $n$、必要換気量 $Q$ の式は

$$ n = \frac{Q}{V} \tag{9}$$

換気回数 $n$ [回/h]、必要換気量 $Q$ [m<sup>3</sup>/h]、室の容積 $V$ [m<sup>3</sup>]

となる。必要換気量と室の容積を割ることにより、1時間に何回換気すれば良いかがわかる。<br>
式3を変形すると換気回数と室の容積をかけることで下式の必要換気量が算出できる。
$$ Q = n \times V$$

### 必要換気量の計算

図4のように室外から室内へ流入する汚染物質量 $C_{0} \cdot Q$、室内で発生する汚染物質量 $M$の和は室内に流入する汚染物質量の総和になる(式4の右辺)。<br>
右辺と室外へ排出される汚染物質量の総和(式10左辺)と等しいとすると
$$ C_{0} \cdot Q + M = C \cdot Q \tag{10}$$
となる。

$C_{0}$ : 室外の汚染物質濃度 [m<sup>3</sup>/m<sup>3</sup>] (1m<sup>3</sup> 中に何m<sup>3</sup> の汚染物質が含まれているのかを表す)、$Q$ : 換気量 [m<sup>3</sup>/h]、$M$ : 室内で発生する汚染物質量 [m<sup>3</sup>/h]、$C$ : 室内の汚染物質濃度[m<sup>3</sup>/m<sup>3</sup>]

<div align="center"><img src="syushi.png" alt="alt テキスト" title="汚染物質収支" width="30%"></div>

$$ 図4　汚染物質量の収支 $$


式10を変形すると
$$ C = C_{0} + \frac{M}{Q}$$
$$ Q = \frac{M}{C-C_{0}} \tag{11}$$


式5となり、室内の汚染物質濃度より必要換気量が計算できる。




### 例題
<br>
居室の二酸化炭素濃度を1000 [ppm] に維持するためには、居住者1人あたりの必要換気量をいくらか。
<br>
人からの二酸化炭素発生量を0.0132 [m<sup>3</sup>/(h・人)]、外気の濃度を350 [ppm]とする。
<br>

許容濃度 $C$ = 1000 [ppm]、外気濃度 $C_{0}$ = 350 [ppm]、二酸化炭素発生量 $M$ = 0.0132 [m<sup>3</sup>/(h・人)]となり、
<br>
図4、式11より

$$ Q = \frac{M}{C-C_{0}} = \frac{0.0132}{(0.001 - 0.00035)} = 20.3 $$

1000ppm = 0.1 % = 0.001 (または ppm に10<sup>-6</sup> 倍する)

よって、必要換気量 Q = 20.3 [m<sup>3</sup>/(h・人)] となる。
